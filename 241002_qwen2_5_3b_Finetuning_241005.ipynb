````python
import torch
#PyTorch 라이브러리를 가져옵니다. 딥러닝 모델 구현에 사용됩니다.

from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, Trainer, TrainingArguments, DataCollatorForSeq2Seq
"""
   - Hugging Face의 transformers 라이브러리에서 여러 클래스와 함수를 가져옵니다.
   - `AutoTokenizer`: 텍스트를 토큰화하는 데 사용됩니다.
   - `AutoModelForCausalLM`: 자동 언어 모델 로딩에 사용됩니다.
   - `BitsAndBytesConfig`: 양자화 설정을 위해 사용됩니다.
   - `Trainer`, `TrainingArguments`: 모델 학습을 위한 클래스들입니다.
   - `DataCollatorForSeq2Seq`: 배치 데이터 생성을 위해 사용됩니다.

"""

#8비트 양자화를 위한 라이브러리입니다.
import bitsandbytes as bnb

#Hugging Face 허브에 로그인하기 위한 함수입니다.
from huggingface_hub import login

from peft import (
    LoraConfig,
    get_peft_model,
    prepare_model_for_kbit_training)
"""
   - PEFT(Parameter-Efficient Fine-Tuning) 라이브러리에서 필요한 함수들을 가져옵니다.
   - `LoraConfig`: LoRA(Low-Rank Adaptation) 설정을 위한 클래스입니다.
   - `get_peft_model`: PEFT 모델을 생성하는 함수입니다.
   - `prepare_model_for_kbit_training`: 8비트 훈련을 위해 모델을 준비하는 함수입니다.

이 코드는 대규모 언어 모델을 효율적으로 fine-tuning하기 위한 준비 단계입니다. 특히:
- 8비트 양자화를 사용하여 메모리 효율성을 높이고,
- LoRA와 같은 파라미터 효율적인 fine-tuning 기법을 적용하며,
- Hugging Face의 생태계를 활용하여 모델과 데이터셋을 쉽게 불러오고 관리할 수 있게 합니다.
"""
````

````python
tokenizer = AutoTokenizer.from_pretrained(model_path)
#미리 선언한 qwen 2.5-3b에 특화된 토크나이저를 자동으로 갖고옴
````

````python
#   - BOS(Beginning of Sentence) 토큰의 ID를 가져옵니다.
#   - 문장의 시작을 나타내는 특수 토큰입니다.
bos = tokenizer.bos_token_id

#   - EOS(End of Sentence) 토큰의 ID를 가져옵니다.
#   - 문장의 끝을 나타내는 특수 토큰입니다.
eos = tokenizer.eos_token_id

#   - PAD(Padding) 토큰의 ID를 가져옵니다.
#   - 시퀀스의 길이를 맞추기 위해 사용되는 특수 토큰입니다.
pad = tokenizer.pad_token_id

#   - PAD 토큰 ID를 EOS 토큰 ID로 설정합니다.
#   - 일부 모델에서는 별도의 PAD 토큰이 없어 EOS 토큰을 PAD 토큰으로 사용합니다.
tokenizer.pad_token_id = eos

# - 패딩을 오른쪽에 추가하도록 설정합니다.
#   - 즉, 시퀀스의 끝부분에 패딩이 추가됩니다.
tokenizer.padding_side = "right"

#   - 입력 시퀀스의 최대 길이를 2048 토큰으로 설정합니다.
#   - 이보다 긴 시퀀스는 잘리게 됩니다.
cut_off_len = 2048
# cut_off_len = 1024

# - 검증 세트의 크기를 전체 데이터셋의 0.5%로 설정합니다.
# - 이는 데이터를 훈련 세트와 검증 세트로 나눌 때 사용됩니다.
val_size = 0.005 # 검증세트는 추후 변경 가능
"""
1. 검증 세트 크기 (val_size = 0.005):
   - 0.005는 실제로 0.5%를 의미합니다. 이는 매우 작은 비율입니다.
   - 일반적으로 검증 세트 크기는 10% (0.1) 에서 20% (0.2) 정도가 흔히 사용됩니다.
   - 0.5%는 매우 작아서, 특히 데이터셋이 작다면 검증 세트가 모델의 성능을 제대로 평가하기에 충분하지 않을 수 있습니다.
   - 더 큰 값(예: 0.1 또는 0.2)을 사용하는 것이 일반적으로 권장됩니다.
검증 세트 크기를 조정하고 싶다면, `val_size = 0.1` 또는 `val_size = 0.2`와 같이 더 큰 값으로 변경하는 것이 좋습니다.
"""


"""
이 설정들은 fine-tuning 과정에서 중요한 역할을 합니다:
- 특수 토큰 ID들을 정의하여 모델이 문장의 시작과 끝을 인식할 수 있게 합니다.
- 패딩 방식을 설정하여 배치 처리 시 일관성을 유지합니다.
- 입력 길이를 제한하여 메모리 사용을 관리하고 처리 속도를 향상시킵니다.
- 검증 세트 크기를 정의하여 모델의 성능을 평가할 수 있게 합니다.
"""
````

````python
#data set의 정보가 들어갈때의 프롬프트 형식 지정
#data set의 input이 들어간 경우에는 prompt_input을 사용 없을 경우에는 prompt_no_input을 사용
prompt_input = "아래는 작업을 설명하는 지시 사항과 추가적인 문맥을 제공하는 입력이 함께 제공된 것입니다. 요청을 적절하게 완료하는 응답을 작성하세요.\n\n### 지시사항:\n{instruction}\n\n### 입력:\n{input}\n\n### 응답:\n"
prompt_no_input = "아래는 작업을 설명하는 지시 사항입니다. 요청을 적절하게 완료하는 응답을 작성하세요.\n\n### 지시사항:\n{instruction}\n\n### 응답:\n"
````

````python
train_dataset  = data.map(
    lambda x: {
        'text': (
            f"{prompt_input.format(instruction=x['instruction'], input=x['input'])}{x['output']}{tokenizer.eos_token}"
            if 'input' in x and x['input'].strip() else
            f"{prompt_no_input.format(instruction=x['instruction'])}{x['output']}{tokenizer.eos_token}"
            )
        }
    )
"""
이렇게 처리된 데이터는 모델이 "지시사항(+입력) -> 출력" 형식의 태스크를 학습하는 데 적합한 형태가 됩니다. 모델은 이 형식을 학습하여 나중에 새로운 지시사항이 주어졌을 때 적절한 출력을 생성할 수 있게 됩니다.
"""
````

````python
train_data = train_data.map(lambda samples: tokenizer(samples["text"]), batched=True)
"""
1. 데이터 전처리: 이 단계는 원시 텍스트를 모델이 이해할 수 있는 숫자 형태로 변환합니다.

2. 메모리 효율성: `batched=True`를 사용함으로써 큰 데이터셋을 효율적으로 처리할 수 있습니다.

3. 인코딩: 이 과정은 텍스트를 모델의 입력 형식에 맞게 인코딩합니다.

4. 데이터셋 변형: 원본 데이터셋에 새로운 필드들(예: `input_ids`, `attention_mask`)이 추가됩니다.

5. 모델 준비: 이렇게 처리된 데이터는 이제 모델 훈련에 직접 사용될 수 있는 형태가 됩니다.
"""
````

````python
print(train_data)
"""
새로운 특성(features) 추가:
   - `input_ids`: 텍스트를 토큰 ID의 시퀀스로 변환한 결과입니다.
   - `attention_mask`: 각 토큰이 실제 입력인지 패딩인지를 나타내는 마스크입니다.

이러한 변경이 의미하는 바:

1. 토큰화 완료: 각 텍스트 샘플이 성공적으로 토큰화되었습니다.

2. 모델 입력 준비: `input_ids`와 `attention_mask`는 모델에 직접 입력될 수 있는 형태입니다.

3. 원본 데이터 보존: 원본 특성들이 유지되어 필요시 참조할 수 있습니다.

4. 처리 효율성: `batched=True` 옵션 덕분에 전체 데이터셋이 효율적으로 처리되었습니다.

이제 이 데이터셋은 모델 훈련에 바로 사용될 수 있는 형태가 되었습니다. `input_ids`와 `attention_mask`는 모델에 입력으로 제공되며, 다른 특성들은 필요에 따라 추가적인 정보나 평가에 사용될 수 있습니다.
"""
````

````python
train_data = train_data.map(
    lambda samples: {
        'input_ids': samples['input_ids'],
        'attention_mask': samples['attention_mask'],
        'labels': samples['input_ids']  # labels 필드 추가
    },
    batched=True
)
"""
이해 하기로는 원래 정보
'input', 'output', 'instruction', 'data_source', 'text'를 이용하여 파인튜닝을 위해
input_ids', 'attention_mask' 를 추가
다시 'labels'도 추가한걸로보임

input_ids', 'attention_mask 의 역할

1. input_ids:
   - 목적: 텍스트를 모델이 이해할 수 있는 숫자 형태로 변환합니다.
   - 과정: 각 단어나 부분 단어(서브워드)를 고유한 정수 ID로 매핑합니다.
   - 필요성: 신경망은 텍스트를 직접 처리할 수 없어 숫자로 변환해야 합니다.

2. attention_mask:
   - 목적: 실제 입력 토큰과 패딩 토큰을 구분합니다.
   - 구성: 실제 토큰은 1, 패딩 토큰은 0으로 표시됩니다.
   - 필요성:
     - 배치 처리 시 서로 다른 길이의 시퀀스를 동일한 길이로 맞추기 위해 패딩을 사용합니다.
     - 모델이 패딩된 부분을 무시하고 실제 입력에만 집중하도록 합니다.

이들이 추가된 주요 이유:

1. 모델 입력 준비:
   - 트랜스포머 기반 모델들은 이러한 형식의 입력을 기대합니다.

2. 효율적인 처리:
   - 정수 ID와 마스크는 텍스트보다 컴퓨터가 효율적으로 처리할 수 있습니다.

3. 배치 처리 지원:
   - 다양한 길이의 입력을 균일한 형태로 처리할 수 있게 합니다.

4. 토크나이저와의 연계:
   - 토크나이저가 자동으로 이러한 변환을 수행하여 모델에 적합한 입력을 생성합니다.

5. 성능 최적화:
   - 이러한 전처리된 형태는 모델의 학습과 추론 속도를 향상시킵니다.



labels의 역할

1. 학습 목표 정의:
   - 'labels'는 모델이 예측해야 할 목표값을 나타냅니다.
   - 언어 모델의 경우, 일반적으로 다음 토큰을 예측하는 것이 목표입니다.

2. 손실 계산:
   - 모델의 예측과 실제 'labels'를 비교하여 손실(loss)을 계산합니다.
   - 이 손실을 바탕으로 모델의 파라미터를 업데이트합니다.

3. 자기회귀 학습:
   - 언어 모델에서 'labels'는 보통 'input_ids'와 동일합니다.
   - 모델은 각 위치에서 다음 토큰을 예측하도록 학습됩니다.

4. 성능 평가:
   - 'labels'를 사용하여 모델의 예측 정확도를 평가할 수 있습니다.

5. 특수 토큰 처리:
   - 패딩이나 마스킹된 토큰에 대해서는 손실을 계산하지 않도록 할 수 있습니다.

6. 다양한 학습 전략:
   - 예를 들어, 특정 부분만 예측하도록 하거나(masked language modeling),
   - 전체 시퀀스를 생성하도록(causal language modeling) 하는 등 다양한 학습 전략을 구현할 수 있습니다.

7. 프레임워크 호환성:
   - 많은 딥러닝 프레임워크와 라이브러리(예: Hugging Face의 Transformers)가 'labels' 필드를 기대합니다.

예를 들어, 'input_ids'가 [1, 2, 3, 4, 5]라면, 'labels'도 [1, 2, 3, 4, 5]가 될 수 있습니다. 모델은 1을 보고 2를 예측하고, [1, 2]를 보고 3을 예측하는 식으로 학습합니다.

따라서 'labels'는 모델이 무엇을 학습해야 하는지, 그리고 얼마나 잘 학습하고 있는지를 정의하고 평가하는 데 필수적인 요소입니다.
"""
````

````python
quantization_config = BitsAndBytesConfig(
    #
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_storage=torch.bfloat16,
    )

# quantization_config = BitsAndBytesConfig(
#     load_in_4bit=True,
#     bnb_4bit_use_double_quant=True,
#     bnb_4bit_quant_type="nf4",
#     bnb_4bit_compute_dtype=torch.float16,  # float16으로 변경
#     bnb_4bit_quant_storage=torch.float16,  # float16으로 변경
# )
"""
이 코드는 BitsAndBytes 라이브러리를 사용하여 모델 양자화(quantization)를 위한 설정을 정의합니다. 각 매개변수의 의미와 목적을 설명드리겠습니다:

1. `load_in_4bit=True`:
   - 모델을 4비트 정밀도로 로드합니다.
   - 메모리 사용량을 크게 줄이면서 성능 손실을 최소화합니다.

2. `bnb_4bit_use_double_quant=True`:
   - 이중 양자화(double quantization)를 활성화합니다.
   - 추가적인 메모리 절약을 제공하며, 정확도 손실을 최소화합니다.

3. `bnb_4bit_quant_type="nf4"`:
   - 양자화 유형을 "NF4"(Normal Float 4)로 설정합니다.
   - NF4는 4비트 양자화에 최적화된 특수한 형식입니다.

4. `bnb_4bit_compute_dtype=torch.bfloat16`:
   - 계산 시 사용할 데이터 타입을 bfloat16으로 설정합니다.
   - bfloat16은 float32와 비슷한 범위를 가지면서 메모리를 절약할 수 있습니다.

5. `bnb_4bit_quant_storage=torch.bfloat16`:
   - 양자화된 가중치를 저장할 때 사용할 데이터 타입을 bfloat16으로 설정합니다.

이 설정의 주요 목적:

1. 메모리 효율성:
   - 4비트 양자화를 통해 모델의 메모리 사용량을 크게 줄입니다.

2. 계산 효율성:
   - bfloat16을 사용하여 계산 속도를 높이면서 정확도를 유지합니다.

3. 대규모 모델 지원:
   - 제한된 하드웨어에서도 대규모 언어 모델을 로드하고 fine-tuning할 수 있게 합니다.

4. 정확도 유지:
   - NF4와 이중 양자화를 통해 정확도 손실을 최소화합니다.

이러한 설정은 특히 대규모 언어 모델을 fine-tuning할 때 유용합니다. 메모리 사용량을 크게 줄이면서도 모델의 성능을 대부분 유지할 수 있어, 제한된 컴퓨팅 리소스로도 고성능 모델을 다룰 수 있게 해줍니다.
"""
````

````python
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    quantization_config = quantization_config,
    torch_dtype = torch.bfloat16,
    device_map = {"" : 0} #colab이나 gpu가 하나면 이렇게 사용
    )
"""
1. `model_path`:
   - 사전 훈련된 모델의 경로 또는 Hugging Face 모델 허브의 모델 식별자입니다.
   - 이전에 정의된 변수로, 특정 모델(예: "Qwen/Qwen2.5-3B")을 가리킵니다.

2. `quantization_config = quantization_config`:
   - 이전에 정의한 양자화 설정을 적용합니다.
   - 4비트 양자화, 이중 양자화 등의 설정이 포함되어 있습니다.

3. `torch_dtype = torch.bfloat16`:
   - 모델의 가중치를 bfloat16 데이터 타입으로 로드합니다.
   - 메모리 사용량을 줄이면서 float32에 가까운 정확도를 유지합니다.

4. `device_map = {"" : 0}`:
   - 모델을 어떤 디바이스에 로드할지 지정합니다.
   - `{"" : 0}`는 전체 모델을 첫 번째 GPU(인덱스 0)에 로드하라는 의미입니다.
"""
````

````python

model.config.use_cache = False
model = prepare_model_for_kbit_training(model)

model.gradient_checkpointing_enable()
"""
1. `model.config.use_cache = False`:
   - 목적: 모델의 캐시 사용을 비활성화합니다.
   - 효과:
     - 메모리 사용량을 줄입니다.
     - 그래디언트 체크포인팅과 호환되도록 합니다.
   - 트레이드오프: 추론 속도가 약간 느려질 수 있지만, 학습 시 메모리 효율성이 향상됩니다.

2. `model = prepare_model_for_kbit_training(model)`:
   - 목적: 모델을 8비트 또는 4비트 훈련에 적합하게 준비합니다.
   - 기능:
     - 특정 레이어(예: 임베딩 레이어)를 전체 정밀도(fp32)로 변환합니다.
     - QLoRA(Quantized Low-Rank Adaptation) 같은 기법을 사용할 때 필요한 설정을 적용합니다.
   - 중요성: 양자화된 모델의 안정적인 fine-tuning을 가능하게 합니다.

3. `model.gradient_checkpointing_enable()`:
   - 목적: 그래디언트 체크포인팅을 활성화합니다.
   - 작동 방식:
     - 순전파 중 일부 중간 활성화를 저장하지 않고, 역전파 중에 필요할 때 재계산합니다.
   - 이점:
     - 메모리 사용량을 크게 줄일 수 있습니다.
     - 더 큰 배치 크기나 더 긴 시퀀스 길이로 학습할 수 있게 합니다.
   - 트레이드오프: 계산 시간이 약간 증가할 수 있습니다.

이 설정들의 종합적인 효과:

1. 메모리 효율성:
   - 캐시 비활성화와 그래디언트 체크포인팅을 통해 메모리 사용량을 크게 줄입니다.

2. 대규모 모델 지원:
   - 제한된 하드웨어에서도 대규모 언어 모델을 fine-tuning할 수 있게 합니다.

3. 양자화 호환성:
   - 4비트 또는 8비트 양자화된 모델의 안정적인 학습을 지원합니다.

4. 학습 안정성:
   - 이러한 최적화 기법들을 통해 대규모 모델의 학습 과정을 더 안정적으로 만듭니다.

이러한 설정은 특히 제한된 GPU 메모리에서 대규모 언어 모델을 fine-tuning할 때 매우 유용합니다. 메모리 사용량을 최적화하면서도 모델의 성능을 유지할 수 있게 해줍니다
"""
````

````python
from peft import LoraConfig, get_peft_model

config = LoraConfig(
    r=8,
    lora_alpha = 16,
    target_modules = ['q_proj', 'k_proj', 'v_proj','o_proj'],
    lora_dropout = 0.05,
    bias = "none",
    task_type = "CAUSAL_LM"
)

model = get_peft_model(model, config)
"""
1. `from peft import LoraConfig, get_peft_model`:
   - PEFT 라이브러리에서 필요한 클래스와 함수를 가져옵니다.

2. `LoraConfig` 설정:
   - `r=8`: LoRA의 랭크를 8로 설정합니다. 이는 적응 행렬의 크기를 결정합니다.
   - `lora_alpha=16`: LoRA 업데이트의 스케일을 조정합니다. 일반적으로 `r`의 2배로 설정합니다.
   - `target_modules=['q_proj', 'k_proj', 'v_proj', 'o_proj']`: LoRA를 적용할 특정 모듈을 지정합니다. 여기서는 attention 모듈의 프로젝션 레이어들을 타겟으로 합니다.
   - `lora_dropout=0.05`: LoRA 레이어에 적용할 드롭아웃 비율입니다.
   - `bias="none"`: 바이어스 파라미터에 LoRA를 적용하지 않습니다.
   - `task_type="CAUSAL_LM"`: 작업 유형을 인과적 언어 모델링으로 지정합니다.

3. `model = get_peft_model(model, config)`:
   - 기존 모델에 LoRA 설정을 적용하여 PEFT 모델을 생성합니다.

LoRA의 주요 이점과 특징:

1. 파라미터 효율성:
   - 전체 모델 파라미터의 작은 부분만 학습합니다.
   - 메모리 사용량과 학습 시간을 크게 줄일 수 있습니다.

2. 성능 유지:
   - 적은 수의 파라미터만 학습하면서도 전체 fine-tuning에 근접한 성능을 달성할 수 있습니다.

3. 유연성:
   - 다양한 downstream 태스크에 쉽게 적용할 수 있습니다.
   - 여러 LoRA 모듈을 독립적으로 학습하고 결합할 수 있습니다.

4. 원본 모델 보존:
   - 원본 사전 훈련 모델의 가중치를 변경하지 않고 유지합니다.

5. 적응성:
   - `target_modules`를 통해 특정 레이어에만 LoRA를 적용할 수 있어 모델 구조에 따라 최적화가 가능합니다.
"""
````

````python
#Hyper parameter setting

output_dir='./easy_train'


# num_epochs = 1 #원본
num_epochs = 2

#micro_batch_size = 1 # 원본
micro_batch_size = 2
#gradient_accumulation_steps =1# 원본
gradient_accumulation_steps =4
warmup_steps = 10
learning_rate = 1e-7
group_by_length = False
optimizer = 'paged_adamw_8bit'

# adam 활용시
beta1 = 0.9
beta2 = 0.95

lr_scheduler = 'cosine'

use_wandb = True
wandb_run_name = 'Single_GPU_Optim'

# use_fp16 = False
# use_bf_16 = True

use_fp16 = True
use_bf_16 = False

evaluation_strategy = 'steps'
eval_steps = 20
save_steps = 20
save_strategy = 'steps'
"""
1. 에폭 (Epochs):
   - 현재 설정: `num_epochs = 1`
   - 추천: `num_epochs = 3` 또는 `num_epochs = 5`
   - 이유:
     - 1 에폭은 모델이 데이터셋을 충분히 학습하기에 부족할 수 있습니다.
     - 3-5 에폭은 모델이 데이터를 더 잘 학습하면서도 Colab의 시간 제한(12시간) 내에 완료될 가능성이 높습니다.

2. 배치 사이즈 (Batch Size):
   - 현재 설정: `micro_batch_size = ` (미지정)
   - 추천: `micro_batch_size = 4` 또는 `micro_batch_size = 8`
   - 이유:
     - GPU RAM이 15GB이므로, 중간 크기의 언어 모델을 고려할 때 4나 8이 적당할 것 같습니다.
     - 더 큰 배치 사이즈는 학습을 안정화시키고 속도를 높일 수 있지만, GPU 메모리 제한을 고려해야 합니다.

3. 그래디언트 누적 (Gradient Accumulation):
   - 현재 설정: `gradient_accumulation_steps = 1`
   - 추천: `gradient_accumulation_steps = 4` 또는 `gradient_accumulation_steps = 8`
   - 이유:
     - 이를 통해 효과적인 배치 사이즈를 늘릴 수 있습니다.
     - 예: `micro_batch_size = 4`와 `gradient_accumulation_steps = 4`를 사용하면 효과적인 배치 사이즈는 16이 됩니다.

4. 평가 및 저장 빈도:
   - 현재 설정: `eval_steps = 20`, `save_steps = 20`
   - 추천: `eval_steps = 100`, `save_steps = 100`
   - 이유:
     - 더 자주 평가하고 저장하면 시간이 많이 소요될 수 있습니다.
     - 100 스텝마다 평가하고 저장하면 진행 상황을 충분히 모니터링하면서도 훈련 속도를 유지할 수 있습니다.
"""
````

````python
trainer = Trainer(
    model=model,

    train_dataset=train_data,
    eval_dataset=val_data,
    args =
    TrainingArguments(
        max_steps=100,
        per_device_train_batch_size = micro_batch_size,
        gradient_accumulation_steps = gradient_accumulation_steps,
        per_device_eval_batch_size = micro_batch_size,
        warmup_steps = warmup_steps,
        num_train_epochs = num_epochs,
        learning_rate = learning_rate,
        adam_beta1 = beta1, # adam 활용할때 사용
        adam_beta2 = beta2, # adam 활용할때 사용
        fp16 = use_fp16,
        bf16 = use_bf_16,
        logging_steps = 1,
        optim = optimizer,
        eval_strategy = evaluation_strategy if val_size > 0 else "no",
        save_strategy="steps",  #스텝기준으로 save
        eval_steps = eval_steps if val_size > 0 else None,
        save_steps = save_steps,
        lr_scheduler_type=lr_scheduler,
        output_dir = output_dir,
        load_best_model_at_end = True if val_size > 0 else False ,
        group_by_length=group_by_length,
        report_to="wandb" if use_wandb else None,
        run_name=wandb_run_name if use_wandb else None,

        ),
    data_collator=DataCollatorForSeq2Seq(tokenizer,padding=True),
    )
````

````python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch

# 8비트 양자화 설정
quantization_config = BitsAndBytesConfig(load_in_8bit=True)

# 모델과 토크나이저 로드
model_name = "hwanni/qwen2.5_3b_Finetunning_241005"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=quantization_config,
    device_map="auto",
    torch_dtype=torch.float16,  # 추가: float16 정밀도 사용
    low_cpu_mem_usage=True  # 추가: CPU 메모리 사용 최적화
)

# 그래디언트 체크포인팅은 추론 시에는 필요 없으므로 제거

# 배치 처리를 위한 함수 정의
def generate_responses(prompts, max_new_tokens=30):
    inputs = tokenizer(prompts, return_tensors="pt", padding=True, truncation=True, max_length=512).to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=max_new_tokens, num_return_sequences=1)
    responses = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    return responses

# 테스트 프롬프트
test_prompts = [
    "인공지능의 미래에 대해 어떻게 생각하시나요?",
    "한국의 전통 음식 중 가장 유명한 것은 무엇인가요?",
    "지구 온난화를 막기 위해 우리가 할 수 있는 일은 무엇일까요?",
    "안녕하세요, 오늘의 날씨는 어떤가요?",
]

print("모델 테스트 결과:")
print("-" * 50)

# 배치 처리
responses = generate_responses(test_prompts)

for prompt, response in zip(test_prompts, responses):
    print(f"입력: {prompt}")
    print(f"출력: {response}")
    print("-" * 50)
````
